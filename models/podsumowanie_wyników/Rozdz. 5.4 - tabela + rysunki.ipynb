```python
import pandas as pd

MODELS = ["Logistic Regression","CART","Random Forest","XGBoost","LightGBM","ANN"]

rows = []
for m in MODELS:
    cv  = pd.read_csv(ART[m]["cv_mean"]).set_axis(["metric","value"], axis=1, copy=False)
    oot = pd.read_csv(ART[m]["oot"]).set_axis(["metric","value"], axis=1, copy=False)

    cv_w  = cv.set_index("metric")["value"]
    oot_w = oot.set_index("metric")["value"]

    rows.append({
        "Model": m,
        "CV_AUC": cv_w.get("AUC"), "CV_PR_AUC": cv_w.get("PR_AUC"), "CV_KS": cv_w.get("KS"),
        "CV_Brier": cv_w.get("Brier"), "CV_LogLoss": cv_w.get("LogLoss"), "CV_ECE": cv_w.get("ECE"),
        "OOT_AUC": oot_w.get("AUC"), "OOT_PR_AUC": oot_w.get("PR_AUC"), "OOT_KS": oot_w.get("KS"),
        "OOT_Brier": oot_w.get("Brier"), "OOT_LogLoss": oot_w.get("LogLoss"), "OOT_ECE": oot_w.get("ECE"),
    })

tab54 = pd.DataFrame(rows).set_index("Model")

# ΔAUC względem Logitu
baseline_auc = tab54.loc["Logistic Regression","OOT_AUC"]
tab54["ΔAUC_vs_Logit_OOT"] = tab54["OOT_AUC"] - baseline_auc

tab54.round(4)
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


# mapowanie nazw kolumn → standard
CV_COLS  = {"AUC":"CV_AUC","PR_AUC":"CV_PR_AUC","KS":"CV_KS",
            "Brier":"CV_Brier","LogLoss":"CV_LogLoss","ECE":"CV_ECE"}
OOT_COLS = {"AUC":"OOT_AUC","PR_AUC":"OOT_PR_AUC","KS":"OOT_KS",
            "Brier":"OOT_Brier","LogLoss":"OOT_LogLoss","ECE":"OOT_ECE"}

rows = []
present = {}

for name, spec in ART.items():
    d = spec["dir"]
    f_cv  = d / spec["cv_mean"]
    f_oot = d / spec["oot"]
    f_roc = d / spec["roc_oot"]
    ok = []
    if f_cv.exists():
        cv = pd.read_csv(f_cv, header=None, names=["metric","value"])
        cv = cv.set_index("metric")["value"]
        for k_new, k_old in CV_COLS.items():
            ok.append(k_old)
        # weź tylko znane metryki
        cv_std = {CV_COLS[k]: float(cv[k]) for k in CV_COLS if k in cv.index}
    else:
        cv_std = {}
    if f_oot.exists():
        oot = pd.read_csv(f_oot, header=None, names=["metric","value"])
        oot = oot.set_index("metric")["value"]
        oot_std = {OOT_COLS[k]: float(oot[k]) for k in OOT_COLS if k in oot.index}
    else:
        oot_std = {}

    if cv_std or oot_std:
        row = {"model": name, **cv_std, **oot_std}
        rows.append(row)
        present[name] = {"dir": d, "roc_oot": f_roc.name}

tab54 = (pd.DataFrame(rows)
           .set_index("model")
           .sort_index())

if tab54.empty:
    raise RuntimeError("Nie udało się zbudować tab54 — sprawdź, czy pliki cv_metrics_mean*.csv / oot_metrics*.csv istnieją.")

display(tab54.round(4))

# ====== 2) Wybór modelu produkcyjnego (tylko XGB/LGBM) + baseline (Logit) ======
cands = [m for m in ["XGBoost", "LightGBM"] if m in tab54.index]
assert len(cands) >= 1, "Brakuje metryk dla XGBoost/LightGBM w tab54."

PROD = max(cands, key=lambda m: tab54.loc[m, "OOT_AUC"])
BASE = "Logistic Regression"
assert BASE in tab54.index, "Brakuje metryk dla Logistic Regression (baseline)."

delta_auc = tab54.loc[PROD, "OOT_AUC"] - tab54.loc[BASE, "OOT_AUC"]
print(f"Model produkcyjny: {PROD} | ΔAUC({PROD} vs Logit, OOT) = {delta_auc:.4f}")

# ====== 3) Rys. 5.4.a — ROC (OOT): PROD vs Logit ======
def show_png(ax, folder: Path, filename: str, title: str):
    p = folder / filename
    if not p.exists():
        raise FileNotFoundError(f"Brak pliku obrazu: {p}")
    ax.imshow(Image.open(p))
    ax.axis("off")
    ax.set_title(title)

fig, axes = plt.subplots(1, 2, figsize=(8,4))
show_png(axes[0], ART[PROD]["dir"], ART[PROD]["roc_oot"],
         f"{PROD}\nAUC_OOT={tab54.loc[PROD,'OOT_AUC']:.3f}")
show_png(axes[1], ART[BASE]["dir"], ART[BASE]["roc_oot"],
         f"{BASE}\nAUC_OOT={tab54.loc[BASE,'OOT_AUC']:.3f}")

fig.suptitle("Krzywe ROC (OOT): model produkcyjny vs logit", y=1.02)
plt.tight_layout()
plt.savefig(OUT / "Rys_ROC_prod_vs_logit.png", dpi=200, bbox_inches="tight")
plt.show()
print(f"ΔAUC({PROD} vs Logit) = {delta_auc:.4f} (zgodnie z Tabelą 5.4)")

# ====== 4) Rys. 5.4.b — Kalibracja (OOT, po izotoniku): Brier i ECE (PROD vs Logit) ======
brier_ece = (
    tab54.loc[[PROD, BASE], ["OOT_Brier", "OOT_ECE"]]
         .rename(columns={"OOT_Brier":"Brier", "OOT_ECE":"ECE"})
)

ax = brier_ece.plot(kind="bar", figsize=(6,4))
ax.set_ylabel("Wartość (mniej = lepiej)")
ax.set_title("Kalibracja (OOT, po izotoniku): Brier i ECE")
ax.set_xticklabels([PROD, BASE], rotation=0)
plt.tight_layout()
plt.savefig(OUT / "Rys_Brier_ECE_prod_vs_logit.png", dpi=200)
plt.show()

display(brier_ece.round(4))
```
